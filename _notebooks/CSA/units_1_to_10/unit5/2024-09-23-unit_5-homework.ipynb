## For 0.90+
Dr. Nefario is trying to assign a recovery time for each minion! Minions who were purple and got cured are very tired, and so are a lot of minions with low energy levels. Create a simple algorithm to calculate how long each minion needs to recover based on their color and energy levels.


```java
import java.util.ArrayList;
import java.util.List;

public class Minion {
    // Instance variables
    private String color;
    private String name;
    private int energyLevel;
    private String gadget;
    private String hair;
    private double height;

    // Static variable to track total energy levels and count of minions
    private static int totalEnergyLevel = 0;
    private static int minionCount = 0;

    // Default constructor
    public Minion() {
        this.color = "Yellow"; // Default color
        this.name = "Unnamed Minion"; // Default name
        this.energyLevel = 50; // Default energy level
        this.gadget = "Banana"; // Default gadget
        this.hair = "Short"; // Default hair style
        this.height = 1.5; // Default height in meters

        // Update static variables
        totalEnergyLevel += this.energyLevel;
        minionCount++;
    }

    // Parameterized constructor
    public Minion(String color, String name, int energyLevel, String gadget, String hair, double height) {
        this.color = color;
        this.name = name;
        this.energyLevel = energyLevel;
        this.gadget = gadget;
        this.hair = hair;
        this.height = height;

        // Update static variables
        totalEnergyLevel += this.energyLevel;
        minionCount++;
    }

    // Method to print minion details
    public void printMinionDetails() {
        System.out.println("Color: " + color);
        System.out.println("Name: " + name);
        System.out.println("Energy Level: " + energyLevel);
        System.out.println("Gadget: " + gadget);
        System.out.println("Hair: " + hair);
        System.out.println("Height: " + height + "m");
        System.out.println();
    }

    // Static method to calculate average energy level
    public static double getAverageEnergyLevel() {
        if (minionCount == 0) return 0; // Prevent division by zero
        return (double) totalEnergyLevel / minionCount;
    }

    // Method to calculate recovery time based on color and energy level
    public int calculateRecoveryTime() {
        int recoveryTime = 0;

        // If the minion is purple, set a higher recovery time
        if (color.equalsIgnoreCase("Purple")) {
            recoveryTime += 10; // More recovery time for purple minions
        }

        // Add recovery time based on energy level
        if (energyLevel < 30) {
            recoveryTime += 10; // Low energy means more recovery time
        } else if (energyLevel < 60) {
            recoveryTime += 5; // Moderate energy means some recovery time
        }

        return recoveryTime; // Return total recovery time
    }

    // Main method to test the Minion class
    public static void main(String[] args) {
        // Create three minions
        Minion minion1 = new Minion(); // Default constructor
        Minion minion2 = new Minion("Purple", "Kevin", 20, "Freeze Ray", "Spiky", 1.4);
        Minion minion3 = new Minion("Yellow", "Stuart", 80, "Banana Blaster", "Short", 1.5);

        // Print minion details
        minion1.printMinionDetails();
        minion2.printMinionDetails();
        minion3.printMinionDetails();

        // Print average energy level
        System.out.println("Average Energy Level of Minions: " + Minion.getAverageEnergyLevel());

        // Calculate and print recovery time for each minion
        System.out.println("Recovery Time for " + minion1.name + ": " + minion1.calculateRecoveryTime() + " hours");
        System.out.println("Recovery Time for " + minion2.name + ": " + minion2.calculateRecoveryTime() + " hours");
        System.out.println("Recovery Time for " + minion3.name + ": " + minion3.calculateRecoveryTime() + " hours");
    }
}

Minion.main(null);
```
